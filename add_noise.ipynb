```python
import librosa
import numpy as np
import soundfile as sf
import os

# Paths
clean_folder = "/Users/taimurulislam/Desktop/MRSP_Seminar/clean_data"
#noisy_folder = "path/to/noisy_data"
os.makedirs("/Users/taimurulislam/Desktop/MRSP_Seminar/__pycache__/noisy_data", exist_ok=True)

# Noise level (adjust for stronger/weaker noise)
noise_factor = 0.02
np.random.seed(42)

# Loop over all clean files
for file_name in os.listdir(clean_folder):
    if file_name.endswith(".wav"):
        clean_path = os.path.join(clean_folder, file_name)
        
        # Load clean audio
        clean_audio, sr = librosa.load(clean_path, sr=None)
        
        # Create white noise
        noise = np.random.randn(len(clean_audio))
        
        # Add noise
        noisy_audio = clean_audio + noise_factor * noise
        
        # Normalize to avoid clipping
        noisy_audio = noisy_audio / np.max(np.abs(noisy_audio))
        
        # Save noisy file
        noisy_path = os.path.join("/Users/taimurulislam/Desktop/MRSP_Seminar/__pycache__/noisy_data", file_name)
        sf.write(noisy_path, noisy_audio, sr)

print("✅ Noisy audio files created in:","/Users/taimurulislam/Desktop/MRSP_Seminar/__pycache__/noisy_data" )
```
